```python
dataset = image_dataset_from_directory(
    r"C:\Users\bilal_gherras\PycharmProjects\Images\content\dataset",
    image_size=(128, 128),
    batch_size=32
)
```

```python
import os
print(os.listdir('/content/dataset'))
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Rescaling(1./255, input_shape=(128, 128, 3)),

    tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(),

    tf.keras.layers.Conv2D(49, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(),

    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(2, activation='softmax')
])
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(dataset, epochs=25)
```

```python
class_names = dataset.class_names
print(class_names)
```

```python
from tensorflow.keras.utils import load_img, img_to_array
import numpy as np

img_path = r"C:\Users\bilal_gherras\Downloads\DATASET\Poire\r2_99_100.jpg"

img = load_img(img_path, target_size=(128, 128))
img_array = img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)

prediction = model.predict(img_array)
print("PrГ©diction brute :", prediction)

classe_predite = class_names[np.argmax(prediction)]
print("Classe prГ©dite :", classe_predite)
print("Confiance :", np.max(prediction) * 100, "%")
```
